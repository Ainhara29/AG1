Respuesta

Primero se crea una lista de numeros y otra lista con los operadores. A su vez creamos nuestra variable target, que se ha definido desde -72 a 80 ya que es el rango donde he visto que existen soluciones, más hallá de esos números, el programa no encuentra ninguna solución. El valor mínimo como target encontrado está al rededor del número -71 y el valor máximo 79.

Por otro lado he definido los bucles de permutaciones: uno para los números con longitud de 5 y otra para los operadores con longitud de 4. Después para cada número y signo he ido guardando la fórmula general y evaluándola con el target. Si la evaluación es la misma que el target, entonces se guarda la solución.

Como podemos observar, nos ha dado una solución de 362.880 posibilidades, como hemos visto en el primer apartado. A su vez, para calcular todas las soluciones, ha requerido de un tiempo de ejecución de alrededor de unos 20 minutos. 

```
from itertools import permutations
numbers = list(map(str, range(1, 10))) 
operators = ["+","-","*","/"]
operators_number = []
target = range(-72, 80)
for tar in target:
    for values in permutations(numbers,len(numbers)-4): 
        for oper in permutations(operators,len(numbers)-5):
            formula = "".join(o+v for o,v in zip([""]+list(oper),values))
            if round(eval(formula)) == tar: 
                print(formula,"=",tar)
                operators_number.append(1)
                print(len(operators_number))
print(len(operators_number))
```

```
import random
random.seed(100)

def seleccion(poblacion, encajar_cromosomas, generacion, target):
    valor_fitness = []
    for cromosoma in poblacion:
        fitness_individual = 100-abs(target-(cromosoma[0]+cromosoma[1]-cromosoma[2]*cromosoma[3]/cromosoma[4]))
        valor_fitness.append(fitness_individual)

    fitness_total = sum(valor_fitness)
    # print('Fitness total: ', fitness_total)
    soluciones = list(zip(valor_fitness,poblacion))
    soluciones.sort(reverse=True)

    for individual in soluciones:
        if individual[0] == 100:
            if individual[1] not in encajar_cromosomas:
                encajar_cromosomas.append(individual[1])
    
    soluciones = soluciones[:4]
    _, poblacion = zip(*soluciones)
    return list(poblacion)

def cruces(poblacion):
    random.shuffle(poblacion)
    cromosoma_padre = poblacion[:2]
    cromosoma_madre = poblacion[2:]
    niño = []
    for i in range(len(cromosoma_padre)):
        cruce=random.randint(1,5)
        fragmentos_padre = [cromosoma_padre[i][:cruce],cromosoma_padre[i][cruce:]]
        fragmentos_madre = [cromosoma_madre[i][:cruce],cromosoma_madre[i][cruce:]]
        firstchild = fragmentos_padre[0]+fragmentos_madre[1]
        niño.append(firstchild)
        secondchild = fragmentos_madre[0] + fragmentos_padre[1]
        niño.append(secondchild)
    return niño

def mutacion(poblacion):
    cromosomas_mutados = []
    for cromosoma in poblacion:
        mutation_site = random.randint(0,4)
        cromosoma[mutation_site] = random.randint(1,9)
        cromosomas_mutados.append(cromosoma)
    return cromosomas_mutados

def cromosomas_intercambiados(generaciones,target):
    poblacion = [[random.randint(1,9) for i in range(5)] for j in range(5)]
    encajar_cromosomas = []
    for generacion in range(generaciones):
        generacion += 1
        # print('Generacion:', generacion)
        poblacion = seleccion(poblacion, encajar_cromosomas, generacion, target)
        cruce_niños = cruces(poblacion)
        poblacion = poblacion + cruce_niños
        poblacion_mutada = mutacion(poblacion)
        poblacion = poblacion + poblacion_mutada
        # random.shuffle(poblacion)

    return encajar_cromosomas

target = range(-72,80)
for tar in target:
    solucion_final = cromosomas_intercambiados(20,tar)
    print('-----------Solucion-----------')
    print(solucion_final)
    print('Tamaño soluciones:  ',len(solucion_final))
    print('Target: ',tar)
```

Por otro lado, como se ha comentado anteriormente, al no existir una fórmula ni máxima ni mínima, se podría encontrar una única solución para cada target. 
Crea un conjunto de todas las operaciones posibles. Y luego solo hace una búsqueda. Sigue siendo una especie de fuerza bruta,  pero te ahorra muchos cálculos repetidos.

```
from itertools import permutations
def find_all_combinations():
    numeros = list(map(str, range(1, 10)))
    output_set = set()
    operadores = ["+","-","*","/"]
    print("Starting Calculations...", end="")
    for values in permutations(numeros,5):
        for oper in permutations(operadores):
            formula = "".join(o + v for o,v in zip([""]+list(oper),values))
            output_set.add(round(eval(formula)))
    return output_set

output = find_all_combinations()

for i in range(-73,81):
    if i in output:
        print(f' {str(i)} Si')
    else:
        print(f' {str(i)} No')
```
